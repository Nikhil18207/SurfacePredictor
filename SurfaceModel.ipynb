```python
import os
import pandas as pd
import numpy as np

folder_path = "D:/SurfaceRecognitionModel/data"  # change if needed
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

cleaned_dfs = []

for file in csv_files:
    try:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)

        # Replace '?' with NaN and drop all-NaN columns
        df.replace("?", np.nan, inplace=True)
        df.dropna(axis=1, how='all', inplace=True)

        # Drop columns like 'Unnamed'
        df = df.loc[:, ~df.columns.str.contains("^Unnamed", na=False)]

        # Keep only numeric columns
        df = df.select_dtypes(include=[np.number])

        # Add default Time column if needed
        if df.shape[1] == 1:
            df.insert(0, "Time", range(1, len(df)+1))

        # Standardize column names
        df.columns = ['Time', 'Resistance'] + list(df.columns[2:])

        # Add surface label from filename
        surface_label = file.replace(".csv", "").split("_")[0].upper()
        df["Surface"] = surface_label

        cleaned_dfs.append(df)

    except Exception as e:
        print(f"‚ùå Error in {file}: {e}")

# Combine all into a master clean dataset
if cleaned_dfs:
    combined_df = pd.concat(cleaned_dfs, ignore_index=True)
    combined_df.to_csv(os.path.join(folder_path, "Cleaned_Surface_Data.csv"), index=False)
    print("‚úÖ Cleaning completed! File saved as 'Cleaned_Surface_Data.csv'")
    print(combined_df.head())

else:
    print("‚ùå No valid CSVs were processed.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Load the cleaned data
df = pd.read_csv("D:/SurfaceRecognitionModel/data/Cleaned_Surface_Data.csv")

# Prepare features and labels
X = df[['Resistance']]  # Only Resistance as input
y = df['Surface']       # Surface label as target

# Encode surface labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)

print("üìä Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
print("üìâ Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# 1Ô∏è‚É£ Load Cleaned Data
df = pd.read_csv("D:/SurfaceRecognitionModel/data/Cleaned_Surface_Data.csv")

# 2Ô∏è‚É£ Remove outliers (very high resistance)
df = df[df['Resistance'] < 50000]

# 3Ô∏è‚É£ Feature Engineering
df['Resistance_diff'] = df['Resistance'].diff().fillna(0)
df['Rolling_mean_5'] = df['Resistance'].rolling(5).mean().fillna(method='bfill')
df['Rolling_std_5'] = df['Resistance'].rolling(5).std().fillna(method='bfill')

# 4Ô∏è‚É£ Prepare Features & Labels
X = df[['Resistance', 'Resistance_diff', 'Rolling_mean_5', 'Rolling_std_5']]
y = df['Surface']
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 5Ô∏è‚É£ Balance Dataset with SMOTE
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(X, y_encoded)

# 6Ô∏è‚É£ Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# 7Ô∏è‚É£ Train XGBoost Classifier
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
model.fit(X_train, y_train)

# 8Ô∏è‚É£ Evaluate
y_pred = model.predict(X_test)

print("‚úÖ Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
print("üìâ Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
